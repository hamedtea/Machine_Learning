An example of linear classification

```python
import matplotlib.pyplot as plt
import numpy as np
# Generate random points for training
np.random.seed(13) # to always get the same points
N = 5
x_h = np.random.normal(1.1,0.3,N)
x_e = np.random.normal(1.9,0.4,N)
plt.plot(x_h,np.zeros([N,1]),'co', label="hobit")
plt.plot(x_e,np.zeros([N,1]),'mo', label="elf")
plt.title('Training samples from two categories')
plt.legend()
plt.xlabel('height [m]')
plt.axis([0.5,2.5,-1.1,+1.1])
plt.show()
```

Nearest neighbor classifier
One of the simplest and most intuitive classifier is the nearest neighbor classifier, or just the
nearest neighbor classification rule, that computes distance to all training samples and then
assigns the sample to be classified the same class the one with the smallest distance
(nearest neighbor)
